```python
import pandas as pd
data = pd.read_csv('C:/Users/Pranav/Downloads/churn-modelling/Churn_Modelling.csv')
data.head()
```

```python
data.drop(['CustomerId', 'RowNumber','Surname'], axis=1, inplace = True)
```

```python
data.head()
```

```python
data.Gender = [1 if each == 'Female' else 0 for each in data.Gender]
data.head()
```

```python
data = pd.get_dummies(prefix = "Geo", data = data, columns = ['Geography'])
data.head()
```

```python
y = data.Exited.values
print(y)
```

```python
x = data.drop(['Exited'], axis = 1)
x.describe()
x.head()
```

```python
import numpy as np
x_norm = (x- np.min(x)) / (np.max(x) - np.min(x))
x_norm.describe()
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtr, xts, ytr, yts = train_test_split(x,y,test_size = 0.3, random_state = 7)
```

```python
print(xtr.shape)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
clf = GaussianNB()
```

```python
clf.fit(xtr, ytr)
```

```python
y_pred = clf.predict(xts)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(yts, y_pred, normalize=True, sample_weight=None)
```

```python
from sklearn.model_selection import cross_val_score
scores= cross_val_score(clf, x, y, cv=5)
print(scores)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=200, random_state=0, max_features=2)
```

```python
from sklearn.model_selection import cross_val_score
scores= cross_val_score(rfc, x, y, cv=5)
print(scores)
```
